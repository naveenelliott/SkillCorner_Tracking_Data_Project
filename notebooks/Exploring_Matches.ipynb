```python
import pandas as pd
import json

base_directory = 'C:/Users/navee/OneDrive/Documents/SkillCorner_Tracking_Data_Project/data/matches'

file_path = f"{base_directory}/1886347/1886347_match.json"

with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
players = pd.json_normalize(data, record_path='players')

players
```

```python
home_team = pd.json_normalize(data['home_team'])
away_team = pd.json_normalize(data['away_team'])
home_kit = pd.json_normalize(data['home_team_kit'])
away_kit = pd.json_normalize(data['away_team_kit'])

teams = pd.concat(
    [home_team.add_prefix('home_'), 
     away_team.add_prefix('away_'), 
     home_kit.add_prefix('homekit_'), 
     away_kit.add_prefix('awaykit_')],
    axis=1
)

teams.drop(columns={'home_short_name', 'away_short_name', 'homekit_id', 'homekit_team_id',
                    'homekit_name', 'homekit_number_color', 'homekit_season.id', 'homekit_season.start_year',
                    'homekit_season.end_year', 'homekit_season.name', 'awaykit_id', 'awaykit_team_id',
                    'awaykit_name', 'awaykit_number_color', 'awaykit_season.id', 'awaykit_season.start_year',
                    'awaykit_season.end_year', 'awaykit_season.name'}, inplace=True)

teams
```
